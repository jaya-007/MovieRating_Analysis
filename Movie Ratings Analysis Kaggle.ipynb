# About the Datasets

* Movies Data :  file contains the data about the movie Id, title and the genre of the movie
* Ratings Data:  file contains the user id, movie id, ratings given by the user and the timestamp of the ratings

```python
# Importing the necessary Python libraries and the movies dataset:
import numpy as np
import pandas as pd
movies = pd.read_csv("/kaggle/input/movie-ratings/movies.dat", delimiter='::', engine='python')
print(movies.head())
```

```python
# Defining the column names:
movies.columns = ["ID", "Title", "Genre"]
print(movies.head())
```

```python
# Importing the ratings dataset:
ratings = pd.read_csv("/kaggle/input/movie-ratings/ratings.dat", delimiter='::', engine='python')
print(ratings.head())
```

```python
# Defining the column names:
ratings.columns = ["User", "ID", "Ratings", "Timestamp"]
print(ratings.head())
```

```python
# Combining the datasets
data = pd.merge(movies, ratings, on=["ID", "ID"])
print(data.head())
```

```python
# Data Visualization
# Distribution of the ratings of all the movies given by the viewers:


ratings = data["Ratings"].value_counts()
numbers = ratings.index
quantity = ratings.values
import plotly.express as px
fig = px.pie(data, values=quantity, names=numbers)
fig.show()
```

```python
# Top 10 movies that got 10 ratings by viewers:

data2 = data.query("Ratings == 10")
print(data2["Title"].value_counts().head(10))
```
